```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="ticks")
```

```python
# Read the dataset
df = pd.read_csv('winemag-data_first150k.csv')
df.head(1)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['country','price'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df=df[['country','points','price']]
```

```python
df.head()
```

```python
df['country'].value_counts(sort=True).head(4)
```

```python
df_top4=df.loc[(df.country=='US')|(df.country=='Italy')|(df.country=='France')|(df.country=='Spain')]
```

```python
sns.catplot(x="country", kind="count", palette="ch:.25", data=df_top4);
```

```python
sns.catplot(y="country", kind="count", palette="ch:.25", data=df_top4);
```

```python
sns.countplot(y="country", data=df_top4, color="c")
```

```python
sns.countplot(x="country", data=df_top4, color="c")
```

### 2. Choose two continuous variables, and plot them three different ways.

```python
sns.lmplot(x="price", y="points", col="country", hue="country", data=df_top4,
          col_wrap=4, ci=None, palette="muted", height=5,
           scatter_kws={"s": 50, "alpha": 1})
```

```python
sns.set(style="darkgrid")
g=sns.catplot(x="points",y="price",hue="country", data=df_top4,
              height=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
```

```python
sns.catplot(x="points", y="price", hue="country", kind="point", data=df_top4)
```

### 3. Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
# histogram, boxplot, pointplot, barplot
```

```python
sns.barplot(x="points",y="country",data=df_top4)
```

```python
sns.catplot(x="country",y="points",data=df_top4)
```

```python
sns.catplot(x="country",y="points",jitter=False,data=df_top4)
```

```python
sns.catplot(x="country",y="points",kind="box",data=df_top4)
```

```python
sns.catplot(x="country",y="points",kind="boxen",
            data=df_top4.sort_values("country"))
```

```python
sns.catplot(x="points", y="country",
            kind="violin", data=df_top4)
```

```python
sns.catplot(x="country", y="points", kind="bar", data=df_top4);
```
